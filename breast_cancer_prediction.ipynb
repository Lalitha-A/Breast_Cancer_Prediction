```python
# Importing necessary Libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Reading the csv file

df = pd.read_csv("/home/mglocadmin/Documents/altimetrix/data.csv")
df
```

```python
df.info()
```

```python
# Dropping the ID column

df.drop('id',axis=1, inplace=True)
```

```python
df['diagnosis'].value_counts()
```

```python
# Plotting the histogram

plt.hist(df['diagnosis'])
plt.xticks([0, 1])
plt.show()
```

```python
df.describe()
```

```python
# Checking the correlation values

df.corr()
```

```python
# Highly correlation features removal 

corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.7)]
df.drop(to_drop, axis=1, inplace=True)
df.shape
```

```python
df.columns
```

```python
# Label encoding of the traget variable

from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df['diagnosis'] = lb.fit_transform(df['diagnosis'])
df
```

```python
# Selection of dependent varaible
y = df[['diagnosis']]
print('Test shape is:',y.shape)

# Selection of Independent variables
X = df.drop(y, axis=1)
print('Train shape is:', X.shape)
```

```python
# Split the data into train and test datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, shuffle=True, random_state=42)
print('Shape of X_train is:{},\n Shape of y_train is: {},\n Shape of X_test is: {},\n Shape of y_test is: {}'.format(X_train.shape,y_train.shape,X_test.shape,y_test.shape))
```

```python
# Logistic Regression Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, confusion_matrix

lr = LogisticRegression()
model = lr.fit(X_train,y_train)
print('Model coefficients are:', model.coef_)
y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
conf = confusion_matrix(y_test,y_pred)
print("Confusion matrix is:", conf)
```

```python
# SVM Classifier

from sklearn.svm import SVC

# Create SVM classifier
svm_classifier = SVC(kernel='linear', C=5.5)

# Train the classifier
svm_classifier.fit(X_train, y_train)

# Predict on test data
y_pred = svm_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
conf = confusion_matrix(y_test,y_pred)
print("Confusion matrix is:", conf)
```

```python
# Naive Bayes classifier

from sklearn.naive_bayes import GaussianNB

# Create Gaussian Naive Bayes classifier
naive_bayes_classifier = GaussianNB()

# Train the classifier
naive_bayes_classifier.fit(X_train, y_train)

# Predict on test data
y_pred = naive_bayes_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
conf = confusion_matrix(y_test,y_pred)
print("Confusion matrix is:", conf)
```

From the above models, SVM model is best for this use case as its recall value is high.
